```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
path_picks = "/Users/theresasawi/Downloads/hypoDD_20161013_20161018_ML.pha"



# Read the text file
with open(path_picks, 'r') as file:
    lines = file.readlines()

# Filter out lines not starting with '#'

event_lines = [line.strip() for line in lines if line.startswith('#')]
# Create pandas DataFrame

df_events = pd.DataFrame([line.split() for line in event_lines])

df_events.rename({0:'#',
                1:'year',
               2:'month',
                3:'day',
                4:'hour',
                5:'minute',
                6:'second',
                7:'lat',
                8:'lon',
                9: 'depth',
                14:'event_id'},
                axis=1,inplace=True)

print(df_events.head(5))

df_events['lon'] = df_events.lon.astype(float)
df_events['lat'] = df_events.lat.astype(float)
lon_ticks = np.linspace(df_events.lon.min(), df_events.lon.max(), num=5)
lat_ticks = np.linspace(df_events.lat.min(), df_events.lat.max(), num=5)
```
